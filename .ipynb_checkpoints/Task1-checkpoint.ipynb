# Which batsmen came to crease but never got a chance to face a ball?

```python
match_groups = del_df.groupby('match_id')
```

```python
match_groups
```

```python
lst=list(del_df['match_id'].unique())
```

```python
import numpy as np
d=dict()
for i in lst:
    grp=match_groups.get_group(i)
    bat=grp['batsman'].unique()
    nstr=grp['non_striker'].unique()
    
    pre=np.setdiff1d(nstr,bat)
    
    if len(pre)!=0:
        d[i]=pre
```

```python
d  # contains dictionary containing key as match_id and value as name of the batsman
```
